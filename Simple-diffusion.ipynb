```python
## (Pseudo) random numbers
# So where are 'real' random numbers?
#seed = 12345  # a nmumber that tells the random number generator ("rng") where/how to start its sequence of "random" numbers
#rng = np.random.default_rng(seed) # shortcut for function
# The sequence is not truly random.  If we use the same seed, we will get the same sequence of numbers.
rng = np.random.default_rng() # shortcut for function
### Toggle with and without seed
rfloat1 = rng.random() # a single random number 
rfloat2 = rng.random() # another 
print(rfloat1,rfloat2) # display the two numbers
```

```python
# Looping to create a list of random numbers
print("The range function generates this array:")
print(list(range(10)))

print ("\nTrial loop")   # \n gives new line
for i in range(10): # remind ourselves how loops work
    print(i)
 
print ("") # skip a line (another way, more clumsy than \n)
print ("array test")
rng_array = []
for i in range(10): # remind ourselves how loops work
    rng_array.append(rng.random())

for x in rng_array:
    print(x)    
```

```python
# Transform a (0,1) random number to a (-1,1) random number
rng_array = [] # create an empty array (list) which we'll fill with (-1,1) random numbers

for i in range(20):
    rng_array.append(2.0*rng.random()-1.0) # need to shift and expand range (or expand, then shift)

for dx in rng_array:
    print(dx)
```

```python
# A histogram is a great way to visualize data
# matplotlib is a library of plotting tools
plt.hist( rng_array )
plt.xlabel("'Random' number")
plt.ylabel("Number of occurrences in array")
```

```python
# Simplest diffusion in one dimension ("1D") using a loop, and plot
#seed = 987654
#rng = np.random.default_rng(seed) # shortcut for function

positions = []
x = 0 # initial position taken to be zero
positions.append(x)
for t in range(200):
    dx = 2.0*rng.random()-1.0 # random step in range (-1,1)
    x = x + dx # new x position taken to be old position plus dx
    positions.append(x) # this appends the *new* position
    
# print(positions)
plt.plot(positions)
plt.xlabel("Time")
plt.ylabel("Position")

# If you re-run this cell, it will produce a different diffusion trajectory every time.  Try it.  How could you prevent this?
```

```python
# Diffusion in two-dimensions (2D)
xposns = [] # can write out as xpositions
yposns = []
x = 0 # initial position taken to be zero
xposns.append(x)
y = 0 # initial position taken to be zero
yposns.append(y)
for t in range(200):
    dx = 2.0*rng.random()-1.0 # random step in range (-1,1)
    x = x + dx # new x position taken to be old position plus dx
    xposns.append(x) # this appends the *new* position
    dy = 2.0*rng.random()-1.0 # random step in range (-1,1)
    y = y + dy # new y position taken to be old position plus dx
    yposns.append(y) # this appends the *new* position
    
# print(xposns)
limit = 15
plt.plot(xposns,yposns)
plt.xlim([-limit, limit])
plt.ylim([-limit, limit])
plt.xlabel("x")
plt.ylabel("y")
```
